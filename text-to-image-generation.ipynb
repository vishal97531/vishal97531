```python

modelid = "CompVis/stable-diffusion-v1-4"
device = "cuda"
pipe = StableDiffusionPipeline.from_pretrained(modelid,revision = "fp16",torch_dtype = torch.float16,use_auth_token = auth_token )
pipe.to(device)
```

```python
def generate():
    with autocast(device):
        image = pipe(input("enter the text:  "),guidance_scale=8.5)["images"][0]
    
    
    img = image.save("/kaggle/working/generatedimage.png")
    image_path = "/kaggle/working/generatedimage.png"
    image = mpimg.imread(image_path)
    plt.imshow(image)
    plt.show()
    
```
